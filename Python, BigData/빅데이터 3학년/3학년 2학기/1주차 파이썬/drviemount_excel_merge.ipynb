```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/aa_Python 프로그래밍 I_II/공공기관_감성분석/data/'
```

```python
tg = pd.read_excel(path + '대구점.xls')
dj = pd.read_excel(path + '대전점.xls')
```

```python
import glob

files  = glob.glob(path + '*.xls')
files
```

```python
for file in files:
    print(file)
    # /data/를 기준으로 텍스트를 분리하고 뒤에 글자
    print(file.split('/data/')[1])
    # /data/를 기준으로 텍스트를 분리하고 뒤에 글자 --> 다시한번 '.'을 기준으로 부리하고 앞에 글자
    print(file.split('/data/')[1].split('.')[0])
    print('*'*100)
```

```python
for file in files:
    print(file)
    print(file.split('/data/')[1])
    print(file.split('/data/')[1].split('.')[0])
    print('*'*100)
```

```python
# 빈 데이터프레임을 만들고 합친다

sf = pd.DataFrame()

for file in files:
    data = pd.read_excel(file) # 파이썬의 판다로 데이터프레임
    name = file.split('/data/')[1].split('.')[0]  # 이름만 추출(parsing)
    data['공공기관'] = name
    sf = sf.append(data)
```
