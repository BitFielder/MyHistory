### 데이터 합치기

```python
# 빈 데이터프레임을 만들고 합친다
import pandas as pd

sf = pd.DataFrame()

for file in files:
    data = pd.read_excel(file) # 파이썬의 판다로 데이터프레임
    name = file.split('.')[0].split('_')[0].split('/')[-1]  # 이름만 추출(parsing)
    data['공공기관'] = name
    sf = sf.append(data)
```
