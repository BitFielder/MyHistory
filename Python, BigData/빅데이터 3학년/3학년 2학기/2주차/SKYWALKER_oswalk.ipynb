>> ### **상대주소(relative address)**
>>> #### **os.walk()를 이용한 방법**
>>> os.walk()는 하위의 폴더들을 for문으로 탐색할 수 있게 해줍니다. 인자로 전달된 path에 대해서 다음 3개의 값이 있는 tuple을 넘겨줍니다.
>>> - root : dir과 files가 있는 path
>>> - dirs : root 아래에 있는 폴더들
>>> -files : root 아래에 있는 파일들

- **os.walk = (경로, 경로 내 디렉토리 리스트, 경로 내 파일 리스트)**
- 즉 tuple안의 원소로 경로를 제외하고 **디렉토리와 파일**은 list 형태

```python
file_path = os.getcwd()
file_path
```

```python
for root in os.walk(file_path):
    print(root)
```

```python
file_path = '/content/sample_data'  # Update the path
for (path, dir, file) in os.walk(file_path):
    print("path:", path)
    print("dir:", dir)
    print("file:", file)
    print("----------")
```
