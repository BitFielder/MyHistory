>>> ##### **Try & Do it**
>>> - **idx, idx_name, notnull데이터프레임 저장**

```python
from tqdm import tqdm
```

```python
not_nulls = []

for file in tqdm(glob(path + '/*.csv')):
    idx = file.replace('.csv','').split('/')[-1].split('_')
    not_nulls.append([idx[0], ' '.join(idx[1:]), pd.read_csv(file, skiprows=3).iloc[:,4:].notnull().sum().max()])
```

```python
not_nulls
```

```python
df = pd.DataFrame(not_nulls, columns=['idx','idx_name','cnt'])
df
```

```python
# indicator 지표(major indicator)
df.idx.str[:2]
```

```python
df['idx_class'] = df.idx.str[:2]
df
```

```python
df['idx_class'].value_counts().sort_index().plot(kind='bar', rot=60, figsize=(12,5))
```

```python
df[df['idx_class'] == 'SE'].sample(10)
```

```python
for text in df[df['idx_class'] == 'SE']['idx_name']:
    print(text)
```

```python
df.query('cnt>=235')
```

```python
for name in df.query('cnt>=235').idx_name:
    print(name)
```

```python
df.query('cnt>=235').to_excel(path + '/_world_bank_quality.xlsx')
```
