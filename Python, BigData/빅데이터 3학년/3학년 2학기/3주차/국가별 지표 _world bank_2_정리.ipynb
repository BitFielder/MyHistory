>>### **3. 선정한 지표에서 값이 없으면 전년도 값으로 채우기**

```python
def get_cleaned_data(data):

    # 연도 관련 컬럼 추출
    cols = pd.Series(data.columns)
    cols = cols[cols.str.contains('\d{4}')].sort_values(ascending=False).values

    # 마지막 연도 + indicator 코드
    last_year = cols[0]
    idx_code = data['Indicator Code'][0]

    # 마지막 연도 기준 데이터 + Nan -> 전년도로 채우기
    for col in cols[1:]:
        data[last_year] = data[last_year].fillna(data[col])

    data = data.rename(columns={last_year:idx_code})

    return data[['Country Code', idx_code]].set_index('Country Code')
```

```python
cleaned = []
for file in indicators.dropna().merge(df, left_on='idx', right_on='idx_name').path:
    cleaned.append(get_cleaned_data(pd.read_csv(file, skiprows=3)))
```

```python
cleaned = []
for file_path in indicators.drop(columns= ['Unnamed: 0']).merge(df, left_on='idx', right_on='idx_name').path:
    cleaned.append(get_cleaned_data(pd.read_csv(file_path, skiprows=3)))
```
