## **회귀분석 Class**

```python
import numpy as np
x = np.array([1,2,4,3,5])
y = np.array([1,3,3,2,5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 4, 3, 5])
y = np.array([1, 3, 3, 2, 5])

# 산점도 그리기
plt.scatter(x, y)
```

```python
# x와 y의 평균을 계산합니다.
mean_x = np.mean(x)
mean_y = np.mean(y)

# 회귀 계수(b1)와 절편(b0)을 계산합니다.
numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)
b1 = numerator / denominator
b0 = mean_y - (b1 * mean_x)

# 회귀식을 출력합니다.
print(f"단순 회귀식: y = {b0:.2f} + {b1:.2f}x")
```

```python
def simple_lr(x, y):
    # x와 y의 평균을 계산합니다.
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    # 회귀 계수(b1)와 절편(b0)을 계산합니다.
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x) ** 2)
    b1 = numerator / denominator
    b0 = mean_y - (b1 * mean_x)

    # 회귀식을 출력합니다.
    #print(f"단순 회귀식: y = {b0:.2f} + {b1:.2f}x")

    return b0, b1
```

```python
b0, b1 = simple_lr(x, y)
```

```python
b1
```

```python
import matplotlib.pyplot as plt
# 예측값을 계산합니다.
x_new = np.linspace(np.min(x), np.max(x), 100)  # 예측을 위한 새로운 x 값들
y_pred = b0 + (b1 * x_new)

# 원본 데이터와 회귀선을 시각화합니다.
plt.scatter(x, y, label='원본 데이터', color='blue')
plt.plot(x_new, y_pred, label='회귀선', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('단순 회귀 분석')
plt.grid(True)
plt.show()
```

```python
 def reg_plot(x, y, b0, b1):
    # 예측값을 계산합니다.
    x_new = np.linspace(np.min(x), np.max(x), 100)  # 예측을 위한 새로운 x 값들
    y_pred = b0 + (b1 * x_new)

    # 원본 데이터와 회귀선을 시각화합니다.
    plt.scatter(x, y, label='원본 데이터', color='blue')
    plt.plot(x_new, y_pred, label='회귀선', color='red')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('단순 회귀 분석')
    plt.grid(True)
    plt.show()
```

### **파일로 만들어 run을 시켜보자**
#### **회귀식을 구하는 함수와 그림을 그리는 함수 2개로 구성된 python 파일**

```python
my_algorithm = '''

import numpy as np
import matplotlib.pyplot as plt

def slr(x, y):

    # x와 y의 평균을 계산합니다.
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    # 회귀 계수(b1)와 절편(b0)을 계산합니다.
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x) ** 2)
    b1 = numerator / denominator
    b0 = mean_y - (b1 * mean_x)

    # 회귀식을 출력합니다.
    print(f"단순 회귀식: y = {b0:.2f} + {b1:.2f}x")
    return b0, b1

def regpl(x, y, b0, b1):

    # 예측값을 계산합니다.
    x_new = np.linspace(np.min(x), np.max(x), 100)  # 예측을 위한 새로운 x 값들
    y_pred = b0 + (b1 * x_new)

    # 원본 데이터와 회귀선을 시각화합니다.
    plt.scatter(x, y, label='원본 데이터', color='blue')
    plt.plot(x_new, y_pred, label='회귀선', color='red')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('단순 회귀 분석')
    plt.grid(True)
    plt.show()
'''
```

```python
# 모듈 파일 작성
with open(os.path.join('/content/my_common', 'lgr.py'), 'w') as file4:
    file4.write(my_algorithm)
```
