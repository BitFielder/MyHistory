```python
# 실행결과 모두 보기
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

# 가로에 모든 결과 화면 보이기 chat GPT에게 
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
import pandas as pd
# 예제 데이터프레임 만들기

def make_df(var, obs):
    data = {c: [str(c) + str(i) for i in obs] 
            for c in var}
    return pd.DataFrame(data, obs)
```

```python
display_side_by_side(
    make_df('ABCD', [0,1,2,3]), make_df('012', ['A','B']), make_df('ACEGIJ', [1,3,5,7,9])
)
```

```python
df1 = make_df('ABCD', [0,1,2,3])
df2 = make_df('DEFG', [1,2,3,4])
display_side_by_side(df1, df2, pd.concat([df1, df1]))
```

# 1. Concat

```python
display_side_by_side(df1, df1, pd.concat([df1, df1])) # 기본은 상하로 합치기(vertical)
```

```python
display_side_by_side(df1, df1, pd.concat([df1, df1], axis=1)) # axis=1, 좌우로 합치기(horizontal)
```

```python
display_side_by_side(df1, df2, pd.concat([df1, df2])) # 기본은 상하로 합치기, 합칠 때 동일한 이름의 컬럼이 없으면 nan
```

```python
display_side_by_side(df1, df2,
    pd.concat([df1, df2], ignore_index=True)
 ) # index를 새롭게 만들기
```

```python
display_side_by_side(df1, df2,
    pd.concat([df1, df2], axis=1)
 ) # 수평으로 합치기
```

```python
display_side_by_side(df1, df2,
    pd.concat([df1, df2], join='inner')
 ) # inner join(기본은 outer join), 서로 컬럼이 있는 경우만 즉 inner는 교집합, outer는 합집합
```

```python
display_side_by_side(df1, df2,
                     pd.concat([df1, df2], axis=1, join='inner')
 ) # 수평으로 서로 인덱스가 경우만
```

```python
display_side_by_side(df1.set_index('D'), df2.set_index('D'),
                     pd.concat([df1.set_index('D'), df2.set_index('D')], axis=1)
) # 수평으로 특정 인덱스를 기준으로
```

# 2. append

```python
display_side_by_side(df1, df1, df1.append(df1))
```

# 3. **merge**
### **<font color='red'> 1:1**
### 동일컬럼 'D'를 기준으로

```python
df3 = pd.merge(df1, df2)
display_side_by_side(df1, df2, df3)
```

### **<font color='red'> 1:M**
### 동일컬럼 'D'를 기준으로

```python
df4 = make_df('DHIJ', [0,1,1,2,2,3])
df5 = pd.merge(df3, df4)
display_side_by_side(df3, df4, df5)
```

### **<font color='red'> M:1**
### 동일컬럼 'D'를 기준으로

```python
display_side_by_side(df4, df3, pd.merge(df4, df3))
#df4.merge(df3)
```

### **<font color='red'> M:M**
### 동일컬럼 'D', 'H'를 기준으로

```python
display_side_by_side(df4, df5, pd.merge(df4, df5))
```

### 왼쪽 파일을 기준으로(how='left') merge하기 때문에 왼쪽 파일의 D컬럼의 'D0'원소값을 포함한다

```python
display_side_by_side(df4, df5, pd.merge(df4, df5, how = 'left'))
```

### merge할 경우 동일한 컬럼명인 경우 앞에 있는 컬럼에_x, 뒤의 컬럼에_y

```python
df6= df5[['A', 'B', 'D', 'H', 'I']]
display_side_by_side(df4,
                     df6,
                     pd.merge(df4, df6, on =['D', 'H'])
)
```

```python
df6= df5[['A', 'B', 'D', 'H', 'I']]
display_side_by_side(df4,
                     df6,
                     pd.merge(df4, df6, on =['D', 'H'], suffixes=('_left', '_right'))
)
```

```python
display_side_by_side(df1,
                     df6,
                     pd.merge(df1, df6, on=['D']) 
                     )# 'D'를 기준으로 서로 합치기
```

```python
display_side_by_side(df1,
                     df6,
                     pd.merge(df1, df6, on=['A'])
 ) # 'A'를 기준으로 서로 합치기
```

```python
display_side_by_side(df1, df6,
                     pd.merge(df1, df6, on=['A'], suffixes=('_left', '_right'))
)
```

### merge할 경우 컬럼명이 달라도 컬럼의 값이 동일하면 merge

```python
df6['J'] = df1['A']; df7 = df6[['H', 'J']]
display_side_by_side(df1, df7,
                     pd.merge(df1, df7, left_on='A', right_on='J') 
                     )# 컬럼명이 틀린 경우에도 서로 이름만 다를 경우
```
